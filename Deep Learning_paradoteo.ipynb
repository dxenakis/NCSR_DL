```python
#Deep Learning 
#1st assignment
#Dimitris Xenakis 
#29/05/2020


import keras
import numpy as np
import keras
from tensorflow.keras import regularizers
import tensorflow as tf
import keras.backend as K
from keras.callbacks import ModelCheckpoint, LearningRateScheduler
from keras.callbacks import ReduceLROnPlateau
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense , Dropout
from keras.optimizers import SGD
import os


#Define learning rate change concerning the epoch iteration
def scheduler(epoch):
    if epoch > 45:
        new_lr =0.0003
        #print("Learning Rate change at epoch:", epoch,"from :",K.get_value(model.optimizer.lr)," to:",new_lr);
        K.set_value(model.optimizer.lr, new_lr) 
    
    
    elif epoch > 32:
        #new_lr = K.get_value(model.optimizer.lr) - 0.001
        #print("Learning Rate change at epoch:", epoch,"from :",K.get_value(model.optimizer.lr)," to:",new_lr);
        new_lr = 0.0005;
        K.set_value(model.optimizer.lr, new_lr) 
    
        
        
    return K.get_value(model.optimizer.lr)


change_lr = LearningRateScheduler(scheduler)

batch_size = 128
num_classes = 10
epochs =120
#load Data
( x_train , y_train ), ( x_test , y_test ) = cifar10.load_data()
#reshape data for network  suitable input_shape
x_train = x_train.reshape(50000 , 3072)
x_test = x_test.reshape(10000 , 3072)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
y_train = keras.utils.to_categorical( y_train , num_classes )
y_test = keras.utils.to_categorical ( y_test , num_classes )




input_shape = x_train.shape[1:]
#Network Declaration and outputs
model = Sequential()
model.add(Dense (120, activation ='relu', input_shape =(3072 ,)))
model.add(Dense( num_classes , activation ='softmax'))
model.summary ()
model.compile (loss='categorical_crossentropy',optimizer =SGD(),
metrics =['accuracy'])
history = model.fit( x_train , y_train ,callbacks=[change_lr],
batch_size = batch_size , epochs =epochs ,verbose =1,validation_data =( x_test , y_test ))
score = model.evaluate ( x_test , y_test , verbose =0)
print ('Test loss:', score [0])
print ('Test accuracy :', score [1])
print('Model Learning rate: ', K.eval(model.optimizer.lr))
```
